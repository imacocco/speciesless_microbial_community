# Ecology without species

Aim: discussing the coexistence of microbial communities without using the concept of species and, instead, starting from a completely disorered system

![clustering of organism](pics/cg.png)

![rabbit](pics/rabbit.png)

This work extends the notion of an “ecomode” to construct a naturally hierarchical, rank-free description of community dynamics in terms of population fluctuation eigenmodes

The approach reduces to the familiar species-based perspective when it applies, but remains well-defined even when the species description breaks down

The basic idea behind this formalism is that species clustering can be reinterpreted as a special form of a more general operation, namely a change of basis in the compositional space.

## definition of the model

Individual rate of abundance change:$\quad r_{\mu} = r_{\mu}(\mathcal{\epsilon},\mathcal{C})$

Interaction matrix: $\quad \mathcal{M}_{\mu\nu} = r_\nu(\mathcal{\epsilon},\mathcal{C}_{2\mu})- r_{\nu}(\mathcal{\epsilon},\mathcal{C})$

Linearised Lotka-Volterra:$\quad \frac{d \delta n_a}{dt}=\sum_{b=1}^K M_{ab}\delta n_b$

indidual-based description: $\quad\mathcal M_{\mu\nu}=\frac{1}{n_b}M_{ab}$

eigenvector of $M_{ab}$ can be converted into an eigenvector of the
matrix $M_{μν}$, by duplicating each entry $n_a$ times. Same eigenvalues!!

resources: $\quad i \in \{1\dots20\}$ with benefit $R_i$

individuals per resource: $\quad T_i = \sum_an_a\sigma_{ai}$

living cost: $\quad \chi_a = \sum_i\sigma_{ai} + \epsilon_a$

resource surplus: $\quad \Delta_a = \sum_i\sigma_{ai}\frac{R_i}{T_i}-\chi_a$

reproduction rate: $\quad r_a = \frac{\Delta_a}{\tau_0\chi_a}$

dynamics: $\quad\frac{d n_a}{dt} = g_a(\vec{n})=\frac{\Delta_a}{\tau_0\chi_a}n_a$

```python
#decide how to evolve the resources across the simulation
res_i = np.logspace(0,3.5,10,base=np.e)
print(res_i)
res_list = []
for i in res_i:
    appo = np.ones(N_resources)
    appo[0]=i
    appo[1]=i
    appo[2]=i
    res_list.append(appo)
res_list = np.array(res_list)
```

```python
n_steps = 7500
pop = Parallel(n_jobs=2)( delayed(third.evolve)(n_steps, temp_resource = res_list[i], traj = 'last') for i in range(len(res_list)) )
```

```python
final_mask = compute_mask(pop,1e-5);
```

```python
fourth = Patch( patch = third , mask = final_mask)
fourth.print_patch()
```

```python
#new equilibration after removing the phenotypes that were not survived with all resources
n_steps = 15000
final_pops = Parallel(n_jobs=2)( delayed(fourth.evolve)(n_steps, temp_resource = res_list[i]) for i in range(len(res_list)) )
```

```python
evl_list = []
evc_list = []
for i in range(len(res_list)):
    M_ab = compute_M_ab(final_pops[i],fourth.phenotypes,fourth.costs,res_list[i])
    evl, evc = linalg.eig(M_ab)
    idx = evl.argsort()
    evl_list.append(evl[idx])
    evc = evc[:,idx]
    evc_list.append(evc.T)
evl_list = np.array(evl_list)
evc_list = np.array(evc_list)
```

As far as we use classical ecological model (LV), based on the definition of species, we will not have a truly gapless spectrum, as a consequence of the partitioning assumption. 
What reasons do we have to belive that natural ecosystems possess this property?
The spectral gap might be an artifact of familiar theoretical models, not necessarely the best approximation of ecological reality.

## Building an ecology without species

+ functional characterisation of phenotypes
+ existance of stable equilibrium
+ set of coexisting phenotypes to be unstructured, so that no grouping is privileged

+ P patches with usual interaction within themselves
+ interaction among patches through redistribution of resources

![multi](pics/multi.png)

single resource availability: $\quad h_i^\alpha=\frac{R_i^\alpha}{T_i^\alpha}$

average resource availabiliy: $\quad h_i^*=\frac{1}{P}\sum_{\alpha=1}^P h_i^\alpha$

resource influx: $\quad R_i^\alpha = \frac{1}{P}R_i-\rho(h_i^\alpha-h_i^*)$

+ aside from the patch structure, no grouping of phenotypes is in any way privileged, offering no possibility for a coarse-grained description
+ point of view of an observer who can measure the community-wide expression of metabolic pathways, but cannot resolve individual patches
+ consider a perturbation where all phenotypes in patch 1 experience an increase in abundance, and all phenotypes in patch 2, a decrease. Such a perturbation will relax quickly, corresponding to one of the dominant (strongly negative) eigenvalues. However, the internal dynamics represented by this mode is effectively hidden from an outside observer whose measurements do not resolve individual patches

with $P\rightarrow\infty$:
+ the number of interacting phenotypes becomes infinite
+ pathway content is unstructured
+ dense spectrum
+ valid ecology at equilibrium
